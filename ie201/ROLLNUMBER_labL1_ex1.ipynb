```python
# ============================================================
# Part 0: Warm-Up Tutorials (NumPy, Pandas, apply, Lambda)
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

# ---- NumPy basics ----
v = np.array([1.0, -2.0, 3.0])
print("Vector:", v)
print("L1 norm:", np.sum(np.abs(v)))
print("L2 norm:", np.sqrt(np.sum(v**2)))

# ---- Pandas basics ----
df_demo = pd.DataFrame({
    "A": [1, 2, np.nan, 4],
    "B": [5, np.nan, 7, 8],
    "Country": ["X", "Y", "Z", "W"]
})

print("\nOriginal DataFrame:\n", df_demo)

df_numeric = df_demo.select_dtypes(include=["int64", "float64"])
df_filled = df_numeric.fillna(df_numeric.median())

print("\nAfter median imputation:\n", df_filled)

# ---- apply() tutorial ----
print("\nColumn-wise mean using apply():")
print(df_filled.apply(lambda col: col.mean(), axis=0))

print("\nRow-wise sum using apply():")
print(df_filled.apply(lambda row: row.sum(), axis=1))

weights = np.array([0.5, -1.0])
print("\nRow-wise dot product (like Xw):")
print(df_filled.apply(lambda row: np.dot(row.values, weights), axis=1))

print("\nLambda (λ) controls regularization:")
print("λ = 0   → No regularization")
print("λ large → Strong regularization\n")
```
